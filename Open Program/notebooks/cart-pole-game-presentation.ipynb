# Define Lots of Functions

```python
def calc_reward(pole_speed, pole_angle):
    
    # declare reward
    reward = 0
    
    # since were dealing with -/+ values we run abs to make it an absolute value. because we dont care if its + or -
    pole_speed = abs(pole_speed)
    pole_angle = abs(pole_angle)
    
    # check what the reward should be and return it
    if pole_speed <= 0.1 and pole_angle <= 0.1:
        reward = 5
    elif pole_speed <= 0.2 and  pole_angle <= 0.1:
        reward = 4
    elif pole_speed <= 0.3 and pole_angle <= 0.1:
        reward = 3
    elif pole_speed <= 0.2 and pole_angle <= 0.2:
        reward = 2
    elif pole_speed <= 0.3 and pole_angle <= 0.2:
        reward = 1
    return reward
```

```python
DECIMALS = 1
TRY_IMPROVE_REWARD = 3

def get_action(state):
    state_but_array = [round(state[0], DECIMALS), round(state[1], DECIMALS), round(state[2], DECIMALS), round(state[3], DECIMALS)]
    for entry in brain_array:
        if entry[0] == state_but_array:
            #print('same state')
            if entry[2] > TRY_IMPROVE_REWARD:
                return entry[1]
            else:
                if entry[1] == 0:
                    return 1
                else:
                    return 0
    return random.choice([0, 1])

def add_brain_cell(state, action, reward):
    # declare i so we can keep track of the current item in the loop and modify it
    i = 0
    for entry in brain_array:
        # for some reason the state is not an array so we make it an array first
        # HEY ITS FUTURE ME, i came back here to round off all numbers to 5 decimals you'll find out later why
        state_but_array = [round(state[0], DECIMALS), round(state[1], DECIMALS), round(state[2], DECIMALS), round(state[3], DECIMALS)]
        if entry[0] == state_but_array:
            print('same')
            if reward > entry[2]:
                print('BETTER REWARD')
                brain_array[i] = [state_but_array, action, reward]
                return True 
            else:
                return True
        i+=1
    print('new')
    brain_array.append([state_but_array, action, reward])   
```

```python
import json

def load_model(model):
    with open(model + '.json', 'r') as f:
        data = json.load(f)
        return data
    
def save_model(model, data):
    save_file = open(model + '.json', "w")  
    json.dump(data, save_file, indent = 6)  
    save_file.close() 
```

```python
def get_avg_score(score_array, max_games):
    total = 0
    for x in score_array:
        total+=x
    return total/max_games
```

```python
def train_and_test_model(attempts, max_score, visuals='on'):
    
    if visuals == 'off':
        env = gym.make("CartPole-v1")
    else:
        env = gym.make("CartPole-v1", render_mode="human") # uncomment this to actually see the progress (slows down the process terribly)
    
    score_array = []
    
    for episode in range(attempts):
        
        # reset all previous made progress in the environment
        env.reset()
        score=0
        
        for frame in range(max_score):
            
            # make the environment visible
            env.render()
            
            # generate random action (0=left or 1=right)
            action = get_action(env.state)
            old_state = env.state
            
            # output of the action
            _, _, game_over, _, _ = env.step(action)
            
            # use the current state values to calulate the reward so we can decide later how good the move was
            reward = calc_reward(env.state[3], env.state[2])
            
            # add values
            add_brain_cell(old_state, action, reward)
            
            # print results
            score+=1
            if game_over:
                break
        score_array.append(score)
        print('total score: ' + str(score))
    print('average score: ' + str(get_avg_score(score_array, attempts)))
    
def test_model(attempts, max_score, visuals='on'):
    
    if visuals == 'off':
        env = gym.make("CartPole-v1")
    else:
        env = gym.make("CartPole-v1", render_mode="human") # uncomment this to actually see the progress (slows down the process terribly)
    
    score_array = []
    
    for episode in range(attempts):
        
        # reset all previous made progress in the environment
        env.reset()
        score=0
        
        for frame in range(max_score):
            
            # make the environment visible
            env.render()
            
            # generate random action (0=left or 1=right)
            action = get_action(env.state)
            old_state = env.state
            
            # output of the action
            _, _, game_over, _, _ = env.step(action)
            
            # use the current state values to calulate the reward so we can decide later how good the move was
            #reward = calc_reward(env.state[3], env.state[2])
            
            # add values
            #add_brain_cell(old_state, action, reward)
            
            # print results
            score+=1
            if game_over:
                break
        score_array.append(score)
        print('score: ' + str(score))
    print('average score: ' + str(get_avg_score(score_array, attempts)))
    print('high score: ' + str(max(score_array)))
```
