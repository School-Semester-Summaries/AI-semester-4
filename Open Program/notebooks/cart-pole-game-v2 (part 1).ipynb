## Data Sourcing
Here's the idea to collect data
1. we have the state of the game
2. we do a random move
3. we get a new state
4. look at the new state and give a reward based on the new state
5. save the old state, action and reward all together

We should have something like this in the end
`brain_array = [state, action, reward]`

So whilst training we keep adding more entries to brain array. We can even overwrite an exact copy of an entry when the reward is better than the reward previous saved. Sounds complicated, let me show you what i mean.

lets say we have this data
```
brain_array = [
1. [[0, 0, 1, 0], "right", 1]
2. [[0, 1, 0, 0], "right", 1]
3. [[1, 0, 1, 0], "left", 3]
]
```
we get a new `entry` whilst training, and this new entry has the same `state` as entry 1 but a better `reward`.
`new_entry = [[0, 0, 1, 0], "left", 5]`

in this case we will overwrite the entry in brain_array. (entry 1 has now been overwritten with a better move)
```
brain_array = [
1. [[0, 0, 1, 0], "left", 5]
2. [[0, 1, 0, 0], "right", 1]
3. [[1, 0, 1, 0], "left", 3]
]
```

# Trying to execute everything i explained

here we declare `brain_array`. It's the array where we save all data in. The idea is to later on per frame check if `brain_array` has already experienced this game state before. If it did happen before, brain_array can return an action for my model to take.

````python
# declare / reset brain_array # we fill the brain array with a irrelevant value since it needs at least one value to not crash later on
brain_array = [[[0.00680, 0.00819, 0.02738, -0.01811], 0, 0]]
````

with the `add_brain_cell` function we can add data to `brain_array`. it also automatically overwrites an entry with the same state that has a better result

**DECIMALS**: the decimal variable is major factor in our learning algorithm. The state actually looks something like this `[0.05365345720, 0.00813546346579, 0.027334573468, -0.01811345735673]`. If we want to test my model, we will check for every frame if the current state has happened before. But if there are 9 decimals this is incredibly specific. With 9 decimals there are a total of 620.000.000.000.000.000.000.000.000.000.000.000.000.000 possible states. So you will never get the same state. That's why we decrease the Decimals for higher chance of encountering a state we have already seen before. 

**5 Decimals**: I tried reducing to 5 decimals but with 15k+ unique states in the brain_array, we never encountered the same state twice. 

**3 Decimals**: Still no duplicate state.

**2 Decimals**: After putting decimals on just 2, we already have a way higher chance of finding states we encountered before. Because 2 decimals leaves us with a total of 25.920.000.000 states that's already way better. Also, we will find the same states more than you think just because some possible states you will never be in. like the cart position is mostly just in the middle of the map leaving out already so much possibilities. I have tried 3 models on 2 decimals, I will show results later.

**1 Decimal**: 1 decimal has only 720.000 possible states. If lots of the possible states are irrelevant. It means we will almost find the same state every frame. So 1 Decimal is definitely the quickest way of training.

````python
# the lower decimals the less precice the agent gets. but lowering this... 
# ...number does speed up training requiring less data so its something you have to weigh a decision of
DECIMALS = 1

def add_brain_cell(state, action, reward):
    # declare i so we can keep track of the current item in the loop and modify it
    i = 0
    for entry in brain_array:
        # for some reason the state is not an array so we make it an array first
        # HEY ITS FUTURE ME, i came back here to round off all numbers to 5 decimals you'll find out later why
        state_but_array = [round(state[0], DECIMALS), round(state[1], DECIMALS), round(state[2], DECIMALS), round(state[3], DECIMALS)]
        if entry[0] == state_but_array:
            #print('same state')
            if reward > entry[2]:
                print('better reward')
                brain_array[i] = [state_but_array, action, reward]
                return True 
            else:
                break
        i+=1
    brain_array.append([state_but_array, action, reward])
                
            
            
````

## Calculating the reward
code for calculating the reward, read the comments for more info, Here we only use the pole speed as input for the reward.

````python
def calc_reward(pole_velocity):
    
    # declare reward
    reward = 0
    
    # since were dealing with -/+ values we run abs to make it an absolute value. because we dont care if its + or -
    pole_velocity = abs(pole_velocity)
    
    # check what the reward should be and return it
    if pole_velocity <= 0.05:
        reward = 6
    elif pole_velocity <= 0.1:
        reward = 5
    elif pole_velocity <= 0.2:
        reward = 4
    elif pole_velocity <= 0.3:
        reward = 3
    elif pole_velocity <= 0.4:
        reward = 2
    elif pole_velocity <= 0.5:
        reward = 1
    return reward
        
        
        
````

## Get best action from brain_array
last thing before we start, we need a function to get the best action. this basically looks in the model brain and looks if the current state has already been solved before, if the result is bigger than 0 it will do the provided action. if its 0 it will do the other action.

**TRY_IMPROVE_REWARD**: this basically means, if the reward currently is for example 3 than we will try another action to see if we can improve 3. The higher this number the bigger the chance of getting better actions, but also the bigger the chance of messing up the game since it might go for worse actions as well. So this variable is my own `Exploration` variable you usually see in RL.

````python
# you can raise the value of this variable to try to improve existing states. the higher the number, the better the chance you will improve an older action. But if you want to go for a highscore might want to consider to turn it off since it can increase the chances on mistakes aswell. You can see it as an exploration variable
TRY_IMPROVE_REWARD = 5 # (bascially the exploration number can be 0-5)

def get_action(state):
    state_but_array = [round(state[0], DECIMALS), round(state[1], DECIMALS), round(state[2], DECIMALS), round(state[3], DECIMALS)]
    for entry in brain_array:
        if entry[0] == state_but_array:
            #print('found state')
            if entry[2] > TRY_IMPROVE_REWARD:
                return entry[1]
            else:
                if entry[1] == 0:
                    return 1
                else:
                    return 0
    return random.choice([0, 1])
````

# Reward Optimalisation
I have observed the gameplay of my model and saw room for improvement. I teached my model how to keep the pole speed as close to 0 as possible. There is one small problem here. If the pole leans to a direction, the speed can still be held close to 0, so the agent keeps the speed at 0 but the pole is slowly falling. to solve this, we can make a new reward system where we can take into account both pole angle and pole speed, but first, lets try only pole angle.

````python
def calc_reward(pole_angle):
    
    # declare reward
    reward = 0
    
    # since were dealing with -/+ values we run abs to make it an absolute value. because we dont care if its + or -
    pole_angle = abs(pole_angle)
    
    # check what the reward should be and return it
    if pole_angle <= 0.1:
        reward = 5
    elif pole_angle <= 0.2:
        reward = 4
    return reward
````

````python
def calc_reward_both(pole_speed, pole_angle):
    
    # declare reward
    reward = 0
    
    # since were dealing with -/+ values we run abs to make it an absolute value. because we dont care if its + or -
    pole_speed = abs(pole_speed)
    pole_angle = abs(pole_angle)
    
    # check what the reward should be and return it
    if pole_speed <= 0.1 and pole_angle <= 0.1:
        reward = 5
    elif pole_speed <= 0.2 and  pole_angle <= 0.1:
        reward = 4
    elif pole_speed <= 0.3 and pole_angle <= 0.1:
        reward = 3
    elif pole_speed <= 0.2 and pole_angle <= 0.2:
        reward = 2
    elif pole_speed <= 0.3 and pole_angle <= 0.2:
        reward = 1
    return reward
````
